```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documents=[
    "I love Dogs",
    "Dogs are a mans best friend",
    "I love Otters",
    "Otters mate for life",
    "Capybaras are the calmest creature",
    "I love me a nice capybara"
    ]
```

```python
vectorizer=TfidfVectorizer() #here bag of words are used from this library
```

```python
x=vectorizer.fit_transform(documents)
```

```python
print("Vocabulary",vectorizer.get_feature_names_out())
```

```python
print("\nBow Matrix:\n",x.toarray()) #has removed words like "I" "a"
```
